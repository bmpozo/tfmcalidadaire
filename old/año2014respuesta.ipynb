```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
datosair = pd.read_csv("historicoaire14.csv",sep=",")
datosair = datosair.drop(datosair.columns[0],axis=1)
datosair.head()
```

```python
calidad14pm10 = []
aux14pm10 = []
calidad14so2 = []
aux14so2 = []
calidad14no2 = []
aux14no2 = []
calidad14co = []
aux14co = []
calidad14o3 = []
aux14o3 = []
for i in range(len(datosair)):
    if (datosair.ix[i,'PM10']<=50):
        calidad14pm10.append('buena')
        aux14pm10.append(0)
    if (datosair.ix[i,'SO2']<=175):
        calidad14so2.append('buena')
        aux14so2.append(0)
    if (datosair.ix[i,'NO2']<=100):
        calidad14no2.append('buena')
        aux14no2.append(0)
    if (datosair.ix[i,'CO']<=5):
        calidad14co.append('buena')
        aux14co.append(0)
    if (datosair.ix[i,'O3']<=90):
        calidad14o3.append('buena')
        aux14o3.append(0)
    if (50<datosair.ix[i,'PM10']<=90):
        calidad14pm10.append('admisible')
        aux14pm10.append(1)
    if (175<datosair.ix[i,'SO2']<=350):
        calidad14so2.append('admisible')
        aux14so2.append(1)
    if (100<datosair.ix[i,'NO2']<=200):
        calidad14no2.append('admisible')
        aux14no2.append(1)
    if (5<datosair.ix[i,'CO']<=10):
        calidad14co.append('admisible')
        aux14co.append(1)
    if (90<datosair.ix[i,'O3']<=180):
        calidad14o3.append('admisible')
        aux14o3.append(1)
    if (90<datosair.ix[i,'PM10']<=150):
        calidad14pm10.append('deficiente')
        aux14pm10.append(2)
    if (350<datosair.ix[i,'SO2']<=525):
        calidad14so2.append('deficiente')
        aux14so2.append(2)
    if (200<datosair.ix[i,'NO2']<=300):
        calidad14no2.append('deficiente')
        aux14no2.append(2)
    if (10<datosair.ix[i,'CO']<=15):
        calidad14co.append('deficiente')
        aux14co.append(2)
    if (180<datosair.ix[i,'O3']<=240):
        calidad14o3.append('deficiente')
        aux14o3.append(2)
    if (150<datosair.ix[i,'PM10']):
        calidad14pm10.append('mala')
        aux14pm10.append(3)
    if (525<datosair.ix[i,'SO2']):
        calidad14so2.append('mala')
        aux14so2.append(3)
    if (300<datosair.ix[i,'NO2']):
        calidad14no2.append('mala')
        aux14no2.append(3)
    if (15<datosair.ix[i,'CO']):
        calidad14co.append('mala')
        aux14co.append(3)
    if (240<datosair.ix[i,'O3']):
        calidad14o3.append('mala')
        aux14o3.append(3)
```

```python
calidad14 = []
for i in range(len(datosair)):
    aux14 = []
    aux14.append(aux14pm10[i])
    aux14.append(aux14so2[i])
    aux14.append(aux14no2[i])
    aux14.append(aux14co[i])
    aux14.append(aux14o3[i])
    maximo = max(aux14)
    if maximo == 0:
        calidad14.append('buena')
    elif maximo == 1:
        calidad14.append('admisible')
    elif maximo == 2:
        calidad14.append('deficiente')
    elif maximo == 3:
        calidad14.append('mala')
    else:
        calidad14.append('no')
```

```python
datosair['calidadpm10'] = calidad14pm10 
datosair['calidadso2'] = calidad14so2 
datosair['calidadno2'] = calidad14no2
datosair['calidadco'] = calidad14co
datosair['calidado3'] = calidad14o3
datosair['calidad'] = calidad14
```

```python
datostiempo = pd.read_csv("historicotiempo14.csv",sep=",")
datostiempo = datostiempo.drop(datostiempo.columns[0],axis=1)
datostiempo.head()
```

```python
datostrafic = pd.read_csv("historicotrafico14.csv",sep=",")
datostrafic = datostrafic.drop(datostrafic.columns[0],axis=1)
datostrafic.head()
```

```python
df = datosair.copy(deep=True)
df['temperatura']=datostiempo['temperatura']
df['tempmax']=datostiempo['tempmax']
df['tempmin']=datostiempo['tempmin']
df['prob_precipitacion']=datostiempo['prob_precipitacion']
df['viento']=datostiempo['viento']
df['intensidad']=datostrafic['intensidad']
df['vmed']=datostrafic['vmed']
df.head()
```

```python
df.to_csv("year2014respuesta.csv",index=False)
```
